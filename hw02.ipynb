```python
import numpy as np
from matplotlib import pyplot as plt
import math
%matplotlib inline
```

```python
count = 1000
length = 1000
theta = 1.0
ks = range(1, 130)
```

```python
def std(get_solution, param: float):
    solutions = np.array([get_solution() for i in range(count)])
    return np.square(solutions - param).sum() / solutions.size
    
```

```python
def uniform(param: float, k: int) -> float:
    xs = np.random.uniform(0, param, length)
    mean = np.power(xs, k).sum() / length
    solution = ((k + 1) * mean)**(1/k)
    return solution
```

```python
def exp(param: float, k: int) -> float:
    xs = np.random.exponential(param, length)
    mean = np.power(xs, k).sum() / length
    solution = (mean / math.factorial(k))**(1/k)
    return solution
```

```python
def plot(get_value, param: float, title: str):
    values = [std(lambda: get_value(param, k), param) for k in ks]
    fig = plt.figure()
    ax = fig.add_axes([0, 0, 1, 1])
    ax.plot(ks, values)
    ax.set_title(title)
    ax.set_xlabel("k")
    ax.set_ylabel("Standard deviation")
```

```python
plot(lambda param, k: uniform(param, k), theta, "Uniform distribution")
```

```python
plot(lambda param, k: exp(param, k), theta, "Exponential distribution")
```
